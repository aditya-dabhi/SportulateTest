```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data2019.csv')
data.head()
```

```python
#Taking the essential features
data = data.iloc[:,[1,2,5,14,16,22,24,25]]
data.head()
```

```python
#Calculating Possessions of aevery team for the whole season
[row, col] = data.shape
for i in range(row):
    data.loc[i,'POS'] = ((data.loc[i,'FGA']+data.loc[i,'TOV'])-(data.loc[i,'ORB']+(0.4*data.loc[i,'FTA'])))
data.head()
```

```python
data = data.iloc[:-1,:]
```

```python
#Finding Offensive and Defensive Efficiency
[row, col] = data.shape
for i in range(row):
    data.loc[i,'OER'] = (data.loc[i,'PTS']/data.loc[i,'POS'])*100
    data.loc[i,'DER'] = (data.loc[i,'OPTS']/data.loc[i,'POS'])*100
```

```python
#Finding League Average
oeravg = data.loc[:,'OER'].mean(axis=0)
deravg = data.loc[:,'DER'].mean(axis=0)
print(oeravg)
print(deravg)
```

```python
#Subtracting efficiency with league average and finding possesions per game
for i in range(row):
    data.loc[i,'Odiff'] = data.loc[i,'OER']-oeravg
    data.loc[i,'Ddiff'] = data.loc[i,'DER']-deravg
    data.loc[i,'Tempo'] = data.loc[i,'POS']/data.loc[i,'G']
```

```python
#Finding league average possessions per game
tempoavg = data.loc[:,'Tempo'].mean(axis=0)
print(tempoavg)
```

```python
#Subtracting possessions per game to league average
for i in range(row):
    data.loc[i,'Tdiff'] = data.loc[i,'Tempo']-tempoavg
data
```

```python
#Predicting scores
def GameSim(team1, team2):
    team1odiff = np.random.normal(data.loc[data['Team']==team1,'OER'],scale=5)-oeravg
    team1ddiff = np.random.normal(data.loc[data['Team']==team1,'DER'],scale=5)-deravg
    team2odiff = np.random.normal(data.loc[data['Team']==team2,'OER'],scale=5)-oeravg
    team2ddiff = np.random.normal(data.loc[data['Team']==team2,'DER'],scale=5)-deravg
    team1expeff = (oeravg + team1odiff + team2ddiff)*(1.01)
    team2expeff = (oeravg + team2odiff + team1ddiff)*(0.99)
    matchtempo = data.loc[data['Team']==team1,'Tdiff'].to_numpy()+data.loc[data['Team']==team2,'Tdiff'].to_numpy()+tempoavg
    team1score = (team1expeff/100)*matchtempo
    team2score = (team2expeff/100)*matchtempo
    score = [team1score, team2score]
    return score
```

```python
#Monte Carlo Simulator 
def simulator(team1, team2):
    team1win = 0
    team2win = 0
    team1score = 0
    team2score = 0
    matchcount=0
    for i in range(5000):
        gm = GameSim(team1,team2)
        if gm[0] > gm[1]:
            team1win+=1
        elif gm[0]<gm[1]:
            team2win+=1
        else:
            matchcount-=1
        team1score+=gm[0]
        team2score+=gm[1]
        matchcount+=1
    #if team1win>team2win:
     #   return 1
    #elif team1win<team2win:
     #   return 0
    print(team1, 'chances of winning = ',(team1win/matchcount)*100,'%')
    print(team2, 'chances of winning = ',(team2win/matchcount)*100,'%')
    print(team1, 'possible score = ',(round((team1score/matchcount)[0])))
    print(team2, 'possible score = ',(round((team2score/matchcount)[0])))
```

```python
#Calling the simulator
#Add the teams from the team list shown in one of the above dataframe
#It takes around 10 to 15 seconds to predict the output
simulator('Los Angeles Lakers', 'Los Angeles Clippers')
```
